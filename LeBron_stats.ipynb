```python
details = details.drop_duplicates(subset=["GAME_ID", "PLAYER_NAME"])
details.head()
```

```python
bron = details[details["PLAYER_NAME"] == "LeBron James"]
bron.drop(["TEAM_ID", "TEAM_CITY", "PLAYER_ID", "PLAYER_NAME", "COMMENT"], axis=1, inplace=True)
bron.head()
```

```python
games_date = games[["GAME_DATE_EST", "GAME_ID", "SEASON"]]
games_date.head()
```

```python
stats = bron.merge(games_date, on="GAME_ID", how="left")
```

```python
seasonal_stats = stats.groupby("SEASON").sum()/stats.groupby("SEASON").count()
```

```python
seasonal_stats.head()
```

# LeBron Win Percentage

```python
bron_details = pd.merge(games, details[details.PLAYER_NAME == "LeBron James"], on = "GAME_ID")
bron_details["home"] = bron_details["TEAM_ID"] == bron_details["TEAM_ID_home"]
bron_details[(bron_details["home"] & (bron_details["PTS_home"] > bron_details["PTS_away"]))]
bron_details["WIN"] = (bron_details["home"] & (bron_details["PTS_home"] > bron_details["PTS_away"])) | ((bron_details["home"] == False) & (bron_details["PTS_home"] < bron_details["PTS_away"]))
```

```python
plt.rcParams.update({'text.color' : "black"})
fig, ax = plt.subplots(figsize=(10,10))
fig.suptitle("LeBron Team", fontsize=20, color = "black")

ax.pie(x = bron_details.groupby(["TEAM_ABBREVIATION"]).count().home, labels = ["CLE", "LAL", "MIA"], autopct="%.1f%%", pctdistance=0.5, wedgeprops=dict(edgecolor='w'))
```

```python
plt.figure(figsize=(10,10))

plt.pie(x=bron_details.groupby("WIN").count().home, labels = ["LOSE", "WIN"], autopct="%.1f%%", pctdistance=0.5, wedgeprops=dict(edgecolor='w'))
plt.title("LeBron Win Percentage", fontsize=20, color = "black")
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.pie(bron_details.groupby(["TEAM_ABBREVIATION"]).count().home, labels = ["CLE", "LAL", "MIA"], radius=1,
       wedgeprops=dict(edgecolor='w'))

ax.pie(bron_details.groupby(["TEAM_ABBREVIATION", "WIN"]).count().home, labels = ["LOSE", "WIN", "LOSE", "WIN", "LOSE", "WIN"], autopct="%.1f%%", radius=0.70,
       wedgeprops=dict(edgecolor='w'))

fig.suptitle("Lebron's W/L Percentage on Each Team", fontsize=20)
plt.show()
```

# Most Successful than any franchise (almost ;))

```python
temp = teams[["TEAM_ID", "ABBREVIATION"]]
```

```python
games["VISITOR_TEAM_WINS"] = (games["HOME_TEAM_WINS"]+1) % 2
games.head()
```

```python
temp = pd.merge(temp, games.groupby("HOME_TEAM_ID").sum()["HOME_TEAM_WINS"], left_on="TEAM_ID", right_index=True)
temp = pd.merge(temp, games.groupby("VISITOR_TEAM_ID").sum()["VISITOR_TEAM_WINS"], left_on="TEAM_ID", right_index=True)
temp["TOTAL_WINS"] = temp["HOME_TEAM_WINS"] + temp["VISITOR_TEAM_WINS"]

team_wins = temp
team_wins = team_wins.append({"ABBREVIATION": "Bron", "TOTAL_WINS":bron_details.groupby("WIN").count().loc[True].max()}, ignore_index=True)
team_wins
```

```python
fig, ax = plt.subplots(figsize=(15,5))
fig.suptitle("TOTAL TEAM WINS SINCE 2003 vs LeBron", color="black", fontsize=18)
sns.barplot(ax=ax, x="ABBREVIATION", y="TOTAL_WINS", data=team_wins.sort_values("TOTAL_WINS", ascending = False))
```
